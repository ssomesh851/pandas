```
#group the dataset by names
names = df.groupby('Name').sum()
names.head()
print(names.shape)
#sort
names.sort_values('Count', ascending=False).head()
```

```
#how many different names exsit in the datset
len(names)
```

```
#what is the name with most occurences
df.Count.idxmax()
```

```
#how many different names have the least occurences
len(names[names.Count == names.Count.min()])
```
