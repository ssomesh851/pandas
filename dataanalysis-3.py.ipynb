```python
#create a lambda function that capitalize string
a = lambda x:x.capitalize()
```

```python
#create mjob, fjob as uppercase
print(df['Mjob'].apply(a).head(5))
print(df['Fjob'].apply(a).head(5))
```

```python
#print the last elements of the dataset
df.tail()
```

```python
df['Mjob'] = df['Mjob'].apply(a)
df['Fjob'] = df['Fjob'].apply(a)
```

```python
#Create a function called majority that return a boolean value to a new column called legal_drinker 
#(Consider majority as older than 17 years old)
def majority(x):
    if x>17:
        return True
    else:
        return False
df['legal_drinker'] = df['age'].apply(majority)
df.head()
```

```python
def times10(x):
    if type(x) is int:
        return 10*x
    return x
df.applymap(times10).head(10)
```
