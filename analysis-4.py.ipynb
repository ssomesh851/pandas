```python
import pandas as pd
import numpy as np
```

```python
#housing market data analysis
#create 3 different series
ser1 = pd.Series(np.random.randint(1, high=5, size=100, dtype='l'))
ser2 = pd.Series(np.random.randint(1,high=4,size=100, dtype='l'))
ser3 = pd.Series(np.random.randint(10000,high=30001, size=100, dtype='l'))
print(ser1, ser2, ser3)
```

```python
#let's create a dataframe by joining the seriess by column
df = pd.concat([ser1,ser2,ser3], axis=1)
df.head()
```

```python
#change the name of the columns to bedrs, bathrs, price_sqr_meter
df.rename(columns= {0:'bedrs', 1:'bathrs', 2:'price_sqr_meter'}).head(10)
```

```python
#create a one column dataframe with the values of the 3 series and assign it to a big column
bigcolumn = pd.concat([ser1,ser2,ser3], axis=0)
#transform it as a dataframe
bigcolumn = pd.DataFrame(bigcolumn)
bigcolumn.head(10)
```

```python
#know the length of the index
len(bigcolumn)
```

```python
#reindex the dataframe so it goes from 0 to 299
bigcolumn.reset_index(drop=True, inplace=True)
bigcolumn
```
